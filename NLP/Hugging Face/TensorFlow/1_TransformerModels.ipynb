Pipeline groups together three steps: preprocessing, passing the inputs through the model, and postprocessing.

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've been waiting for a HuggingFace course my whole life.",
        "I hate this so much!",
    ]
)
```

```python
from transformers import AutoTokenizer
from transformers import TFAutoModel

#Preprocessing must be same way as the model was pretrained, AutoTokenizer.from_pretrained fetchs it
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = TFAutoModel.from_pretrained(checkpoint)

raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]

inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="tf")
outputs = model(inputs)

print(outputs.last_hidden_state.shape)
```


*   Batch size: The number of sequences processed at a time.
*   Sequence length: The length of the numerical representation of the sequence.
*   Hidden size: The vector dimension of each model input.

```python
from transformers import TFAutoModelForSequenceClassification
import tensorflow as tf

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(inputs)

print(outputs.logits.shape)
print(outputs.logits)

predictions = tf.math.softmax(outputs.logits, axis=-1)
print(predictions)
```
